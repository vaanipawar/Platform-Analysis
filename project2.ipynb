```python

df['date'] = pd.to_datetime(df['date'])  
df = df.sort_values(by='date') 
print(df['date'].to_string())
```

average review per month

```python
avg_ratings = df.groupby(df['date'].dt.to_period('M'))['rating'].mean()
min_value = avg_ratings.min()
max_value = avg_ratings.max()
min_month = avg_ratings.idxmin()
max_month = avg_ratings.idxmax()
months = avg_ratings.index.astype(str)

plt.bar(months,avg_ratings)
plt.scatter(min_month.strftime("%Y-%m"), min_value, color="red", s=100, label=f"Min: {min_value:.2f}")
plt.scatter(max_month.strftime("%Y-%m"), max_value, color="green", s=100, label=f"Max: {max_value:.2f}")

plt.xlabel("month")
plt.ylabel("average rating")
plt.xticks(rotation=90) 
plt.legend() 



plt.show()
print(min_month, "-",min_value)
print(max_month, "-",max_value)
```
